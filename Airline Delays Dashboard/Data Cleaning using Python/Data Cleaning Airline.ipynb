```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("DelayedFlights.csv")
df = df.drop(df.columns[0], axis=1)
```

```python
# Sanity Check for Negative values in ArrTime, CRSEArrivalTime, DepTime and CRSEDepTime
print(all(i < 0 for i in df['ArrTime']))
print(all(i < 0 for i in df['CRSArrTime']))
print(all(i < 0 for i in df['DepTime']))
print(all(i < 0 for i in df['CRSDepTime']))
```

```python
df.describe()
```

```python
df.info()
```

```python
# Converting Categorical Variables to object type
df[df.columns[0:4]] = df[df.columns[0:4]].astype(object)
df[df.columns[8:10]] = df[df.columns[8:10]].astype(object)
df[df.columns[21:24]] = df[df.columns[21:24]].astype(object)
```

```python
# Renaming the labels within the variables
df['Cancelled'] = df['Cancelled'].replace(
    [1, 0], ["Cancelled", "Not Cancelled"])
df['Diverted'] = df['Diverted'].replace([1, 0], ["Diverted", "Not Diverted"])
```

```python
print(df['Cancelled'].value_counts())
print(df['Diverted'].value_counts())
```

```python
# Checking for null values
df.isna().sum()
```

```python
df.head(3)
```

# Handling Null Vaules


```python
# Null value Imputation using fillna(0) Method
df['CarrierDelay'] = df['CarrierDelay'].fillna(0.0)
df['WeatherDelay'] = df['WeatherDelay'].fillna(0.0)
df['NASDelay'] = df['NASDelay'].fillna(0.0)
df['SecurityDelay'] = df['SecurityDelay'].fillna(0.0)
df['LateAircraftDelay'] = df['LateAircraftDelay'].fillna(0.0)

df.isna().sum()
```

```python
# Checking value counts for categorical variables

# Filter categorical variables
num_cols = df._get_numeric_data().columns
cols = df.columns
cat_cols = list(set(cols) - set(num_cols))
cat_cols

# Value counts
for col in cat_cols:
    if col in ['DayOfWeek', 'UniqueCarrier', 'Month', 'Cancelled', 'DayofMonth', 'CancellationCode', 'Diverted']:
        print(df[col].value_counts())
```
